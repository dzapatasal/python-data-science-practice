Fórmula matemática: **y = ax + b**
- a : pendiente = y2 - y1 / x2 - x1 (obteniendo dos puntos de la gráfica de izquierda a derecha)

📘 ¿Qué significa cada parte?

- **y**: Es el **valor de salida** o resultado (también llamada *variable dependiente*).
- **x**: Es la **variable de entrada** (también llamada *variable independiente*).
- **a**: Es la **pendiente** de la recta. Indica **cuánto cambia y cuando x aumenta en 1**.
- **b**: Es el **intercepto con el eje y**. Es el punto donde la recta **cruza el eje y (cuando x = 0)**.


```python
# En el ejemplo asignamos:
# 'x' → fechas
# 'a' → pendiente
# 'a' → 0.55 asumimos dos puntos en el plano de la recta de izquierda a derecha

x = fechas
y = 0.5 * x + 78
#(hallamos este valor despejando uno de los puntos asumidos)
y
```

```python
plt.plot(fechas, Moscu)
plt.plot(x,y)
plt.title('Apples: 2013-2016 Moscow City Prices')
plt.xlabel('Total months')
plt.ylabel('Precio (₽‎/.)\nrublo ruso')
plt.legend(['Gráfica lineal','Recta aritmética'])
plt.grid()
```

```python
# Generamos un modelo de regresion
# Explicacion mas detallada en '\notes\terminos.md'
np.linalg.norm(Moscu-y)
```

## Análisis de regresión lineal simple


$$\alpha = pendiente$$
$$\beta = interseccion$$
$$n = nro. observaciones$$
$$y = precio$$
$$ x = fechas $$

La solución para la pendiente es:
$$ \alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

y para la interseccion:
$$ \beta = \frac{\sum y - \alpha \sum x}{n} $$

y obtenemos la ecuación:
$$ y = \alpha x + \beta $$

```python
# ANALISIS DE REGRESION LINEAL SIMPLE
# PARA MAYOR DETALLE
# archivo 'formula_regresion_lineal.md'
# path: notes\

# DEFINIENDO VALORES:
n = np.size(Moscu) # nro. ocurrencias = nr. meses totales
y = Moscu
x = fechas
```

Cálculo de la pendiente
$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2}$$

```python
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2)- np.sum(x)**2)
a
```

Cálculo de la intersección
$$ \beta = \frac{\sum y - \alpha \sum x}{n} $$

```python
b = (np.sum(y) - (a*np.sum(x)))/n
b
```

```python
print(f'Pendiente (a): {a}\nIntersección (b): {b}')
```

Cálculo de Y
$$ y = \alpha x + \beta $$

```python
y = a * x + b
y # 'y' a sido reasignada con los valores predichos por el modelo de regresión lineal
```

Calculo del error promedio con valores de la regresion lineal

```python
calculo_norma = np.linalg.norm(Moscu-y)
error_promedio = calculo_norma/Moscu.shape
print(f'Error promedio: {error_promedio}')
```

El **error promedio** en la regresión lineal es una medida de la precisión de tu modelo. Te dice, en promedio, qué tan lejos están las predicciones de tu modelo de los valores reales que observaste.

En el ejemplo: (error de ~1.17 rublos con precios entre 73-131 soles).
