## Comparación entre *arrays*

```python
Moscu.shape
```

Separación de datos por años: Dividimos un array de precios de manzanas de Moscú en arrays anuales para analizar el comportamiento del precio en diferentes años (2013, 2014, 2015 y 2016).

```python
# AÑO 2013: los 12 primeros meses de la matriz
moscu_1 = Moscu[0:12] # '0' Enero , '12' Diciembre. En realidad es 11 pero el parametro es exclusivo (toma hasta un número antes)

# AÑO 2014: los 12 meses despues de su antecesor en la matriz
moscu_2 = Moscu[12:24]

# AÑO 2015: los 12 meses despues de su antecesor en la matriz
moscu_3 = Moscu[24:36]

# AÑO 2016: los 12 meses despues de su antecesor en la matriz
moscu_4 = Moscu[36:48]
```

```python
# Creacion de periodos de 12 meses por año
plt.plot(np.arange(1,13,1),moscu_1) # Grafica de linea simple del 2013
plt.plot(np.arange(1,13,1),moscu_2) # Grafica de linea simple del 2014
plt.plot(np.arange(1,13,1),moscu_3) # Grafica de linea simple del 2015
plt.plot(np.arange(1,13,1),moscu_4) # Grafica de linea simple del 2016
plt.title('Apples: 2013-2016 Moscow Prices')
plt.xlabel('Meses')
plt.legend(['2013','2014','2015','2016']) # Legenda del gráfico
plt.ylabel('Precio (₽‎/.)\nrublo ruso')
plt.grid()
plt.show()
```

```python
# Comparasion estricta de arrays entre años: 2013 y 2014
np.array_equal(moscu_1, moscu_2)
```

```python
# Esta funcion nos permite apreciar que los valores oscilan entre 77, 78 y 80
np.allclose(moscu_1,moscu_2, 0.1) # Tolerancia: 3º parametro
```

# Tratamiento de NaN ('Not a Number')

```python
# Verificación de precios en Moscú
# Verificación de cuanto a sido el precio de las manzanas a lo largo de los 7 años
np.mean(Moscu)
# np.mean(moscu_1) → # Verificación de precios solo en el año 2013
```

```python
# Verificación de precios en Kakiningrado
# Todo el rango de años (2013-2016) 7 años
np.mean(Kaliningrado)
# error: NaN
```

```python
# Verificamos mediante grafica el error
plt.plot(fechas, Kaliningrado)
# Se visualiza que hay un 'dato vacio o faltante'
```

```python
# Verificaciòn visual de valores 'NaN' mediante funcion 'isnan()'
# Se visualiza un elemento 'True' → elemento NaN
np.isnan(Kaliningrado)
```

```python
# Verificación lógica
# Suma de todos los 'False' ya que cada uno equivale a '0'
# Si existe 1 o mas valores 'True' o NaN el resultado tendrá que ser 1 o más.

np.sum(np.isnan(Kaliningrado))
```

```python
# Para cambiar el número lo -INTERPOLAMOS-
# INTERPOLACION significa "estimar un valor desconocido que se encuentra entre dos o más valores conocidos."
# Promedio de la suma del numero anterior y posterior al NaN
(Kaliningrado[3] + Kaliningrado[5])/2
```
